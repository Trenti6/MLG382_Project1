```python
%matplotlib inline
import pandas as pd
df = pd.read_csv("Student_performance_data .csv")
import os
print(os.listdir())  # This will show you the exact file name in your current directory
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
```

```python
# GPA
sns.histplot(df['GPA'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('GPA Distribution')
```

```python
# Study Time
sns.histplot(df['StudyTimeWeekly'], kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Weekly Study Time Distribution')
```

```python
# Absences
sns.histplot(df['Absences'], kde=True, ax=axes[2], color='lightgreen')
axes[2].set_title('Absences Distribution')
```

```python
plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis: GPA vs. Parental Support and Extracurriculars

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv("Student_performance_data .csv")  

# List files in the current directory
print(os.listdir())  # This will show you the exact file name in your current directory


```

```python
# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
```

```python
# GPA vs Parental Support
sns.boxplot(x='ParentalSupport', y='GPA', hue='ParentalSupport', data=df, ax=axes[0], palette='Blues', legend=False)

axes[0].set_title('GPA vs. Parental Support')
```

```python
# GPA vs Extracurriculars
sns.boxplot(x='Extracurricular', y='GPA', hue='Extracurricular', data=df, ax=axes[1], palette='Purples', legend=False)

axes[1].set_title('GPA vs. Extracurricular Involvement')
```

```python
plt.tight_layout()
plt.show()
```
